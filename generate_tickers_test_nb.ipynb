```python
ticker_aapl = yf.Ticker("AAPL")
info_aapl = ticker_aapl.info
info_aapl
```

```python
epoch_time_aapl = info_aapl.get("firstTradeDateMilliseconds")
if epoch_time_aapl:
    listing_date_aapl = datetime.fromtimestamp(epoch_time_aapl / 1000).strftime(
        "%Y-%m-%d"
    )
    print(listing_date_aapl)
```

```python
from yfinance import Ticker


for symbol in tickers_comb:
    try:
        ticker_obj: Ticker = yf.Ticker(symbol)

        # Fetch the metadata (one quick request)
        info: dict = ticker_obj.info

        # This field provides the first trade date in Unix timestamp (Epoch)
        epoch_time = info.get("firstTradeDateMilliseconds")

        if epoch_time:
            # Convert Unix timestamp to a readable date
            listing_date = datetime.fromtimestamp(epoch_time / 1000).strftime(
                "%Y-%m-%d"
            )
            tickers_comb_dict[symbol] = listing_date
            print(f"{symbol:<8} | {listing_date}")
        else:
            print(f"{symbol:<8} | Metadata field not found")

    except Exception as e:
        print(f"{symbol:<8} | Error: {e}")
```
